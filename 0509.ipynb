```python
from skimage import io,data,img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np

def lsb(x):
    lsb_list = []
    for i in range(0,8):
        lsb_list.append(x%(2**(i+1)))
        
    for j in range(7,0,-1):
        lsb_list[j] = lsb_list[j] - lsb_list[j-1]
        if lsb_list[j] != 0:
            if(lsb_list[j]%2) == 0:
                lsb_list[j] = 1
        else:
            lsb_list[j] = 0
    return lsb_list

img = io.imread('C:/Users/S_H/Documents/Multimedia/圖片檔ver2/ape.jpg')
rows,cols,dims = img.shape
plt.figure(figsize=(16,16))

img_gray = img_as_ubyte(img.copy())
img_sample = img[:,:,0].copy()

for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k]
        img_gray[i,j] = sum/3
        img_sample[i,j] = sum/3
        
        lsb_list = lsb(img_sample[i,j])
        if lsb_list[6] == 1:
            img_sample[i,j] = 255
        else:
            img_sample[i,j] = 0
plt.subplot(3,1,1)
plt.title('Original')
plt.imshow(img)

plt.subplot(3,1,2)
plt.title('Grey')
plt.imshow(img_gray)

plt.subplot(3,1,3)
plt.title('LSB6')
plt.imshow(img_sample)

plt.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import math

def distance(clr1, clr2):
    dif = 0
    for i in range(0,3):
        dif = dif + math.pow((clr1[i]-clr2[i]),2)
    return math.sqrt(dif)

def makeColor(image,clr):
    rows,cols,dims = image.shape
    for i in range(0,rows):
        for j in range(0,cols):
            if distance(image[i,j], clr) < 50.0:
                image[i,j,0] = 255
    return image
img = io.imread('C:/Users/S_H/Documents/Multimedia/圖片檔ver2/ape.jpg')
io.imshow(img)

brown = [45,25,15]
img = makeColor(img, brown)
plt.figure()
io.imshow(img)
io.show()
```
