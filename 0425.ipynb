```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('C:/Users/S_H/Documents/Multimedia/圖片檔ver2/black cat.jpg')
rows,cols,dims = img.shape

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)

img2 = np.uint8(np.zeros((rows,cols,3)))
threshold = 128

for i in range(0,rows):
    for j in range(0,cols):
        if img[i,j,0] > threshold:
            for k in range(0,3):
                img2[i,j,k] = 255
        else:
            for k in range(0,3):
                img2[i,j,k] = 0
print(img2.shape)
plt.subplot(1,2,2)
plt.title('Binimage')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('C:/Users/S_H/Documents/Multimedia/圖片檔ver2/black cat.jpg')
io.imshow(img)

rows,cols,dim = img.shape
half = np.matrix([[0,128],[192,64]])#2x2 dithering
img2 = img[:,:,0].copy()

for i in range(0,rows):
    for j in range(0,cols):
        index_i = i%2
        index_j = j%2
        if img2[i,j] > half[index_i,index_j]:
            img2[i,j] = 255
        else:
            img2[i,j] = 0
plt.figure()
io.imshow(img2)
io.show()
```
