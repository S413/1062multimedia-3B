```python
def factorial2(g):
    if g > 0:
        return g*factorial2(g-1)
    else:
        return 1
    
```

```python
factorial2(e)
```

```python
def fib(n):
    """Print a Fibonacci series up to n"""
    a,b = 0,1
    while a < n:
        print(a, end='')
        a,b = b,a+b
        print()
        
```

```python
fib(2000)
```

```python
from skimage import io,data
import numpy as np

img = io.imread('C:/Users/S_H/Documents/Multimedia/圖片檔ver2/black cat.jpg')
rows,cols,dims = img.shape

row2 = int(rows/2)
col2 = int(rows/2)

img2 = img

for i in range(row2,rows):
    for j in range(0,rows):
        for k in range(0,3):
            img2[i,j,k] = img[row2+(-1)*(i-row2),j,k]
            
io.imshow(img2)
io.show()
```
